```python
import urllib3
urllib3.disable_warnings()
from bs4 import BeautifulSoup
url_root='https://medlineplus.gov/'
str_page=''
links = []
for i in range(ord('a'), ord('w')+1):
    str_page= str(chr(i))
    url=url_root+'healthtopics_{}.html'.format(str_page)
   #url ='https://doctor.webmd.com/results?so=zip=19331'
   #url='https://doctor.webmd.com/hospital/chestnut-hill-hospital-philadelphia-pa-73e9de01-b74c-423f-b0c4-4ed34f314399-physicians#pulmonary-critical-care'
    https = urllib3.PoolManager()
    response = https.request('GET', url)
    page = BeautifulSoup(response.data,'lxml')
    index = page.find("section")
    for h in index.find_all('a', href=True):
        links.append((h["href"],h.text))
url=url_root+'healthtopics_xyz.html'
https = urllib3.PoolManager()
response = https.request('GET', url)
page = BeautifulSoup(response.data,'lxml')
print(url)
for i in range(ord('X'),ord('Z')+1):
    str_page= str(chr(i))
    index = page.find('div',{'id':'section_{}'.format(str_page),'class':'section-body'})
    for h in index.find_all('a', href=True):
        links.append((h["href"],h.text))
print(len(links))
print('\n\n')
print(links)
```
